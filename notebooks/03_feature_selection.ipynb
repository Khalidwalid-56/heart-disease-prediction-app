# 03 - Feature Selection


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load the cleaned dataset before feature selection
df = pd.read_csv("/content/Cleaned_dataset_BeforeFeatureSelection.csv")
```

```python
# Display the correlation matrix using a heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), annot=True, cmap="YlGnBu")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Separate features and target
X = df.drop("target", axis=1)
y = df["target"]
```

```python
# Train a Random Forest classifier to determine feature importance
model = RandomForestClassifier(random_state=42)
model.fit(X, y)
```

```python
# Visualize feature importance
importances = pd.Series(model.feature_importances_, index=X.columns)
importances.sort_values(ascending=True).plot(kind='barh', figsize=(10,6))
plt.title("Feature Importances - Random Forest")
plt.xlabel("Importance Score")
plt.show()
```

```python
# Keep only selected features and target
df1 = df[X.columns]
df1["target"] = df["target"]
```

```python
# Save the dataset after feature selection
df1.to_csv("data\\Cleaned_dataset_AfterFeatureSelection.csv", index=False)
print("\nCleaned dataset saved to data\\Cleaned_dataset_AfterFeatureSelection")
```

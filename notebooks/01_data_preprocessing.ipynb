# 01 - Data Preprocessing


```python
# Importing essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
dataset = pd.read_csv("/content/Heart_Disease_data.csv")
```

```python
# Display basic dataset shape and preview
print("Dataset shape:", dataset.shape)
print("\nFirst 5 rows:")
dataset.head()
```

```python
# Check for nulls and duplicates
print("\nNull values:\n", dataset.isnull().sum())
print("\nDuplicated rows:", dataset.duplicated().sum())
```

```python
# Drop duplicated rows
dataset.drop_duplicates(inplace=True)
print("\nDuplicated rows after dropping:", dataset.duplicated().sum())
```

# Dataset summary


```python
# Summary statistics of dataset
print("\nDataset description:")
dataset.describe()
```

```python
# Info about data types and non-null counts
print("\nDataset info:")
dataset.info()
```

```python
# Save cleaned dataset
dataset.to_csv("data\\Cleaned_dataset_BeforeFeatureSelection.csv", index=False)
print("\nCleaned dataset saved to data\\Cleaned_dataset_BeforeFeatureSelection")
```

# 02 - PCA Analysis


```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Load the cleaned dataset
df = pd.read_csv("/content/Cleaned_dataset_BeforeFeatureSelection.csv")
```

```python
# Separate features and target variable
X = df.drop("target", axis=1)
y = df["target"]
```

```python
# Standardize the features before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply PCA and reduce features to 2 principal components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Show explained variance ratio
print("Explained variance ratio:", pca.explained_variance_ratio_)
```

```python
# Visualize the PCA result
plt.figure(figsize=(10,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='coolwarm', edgecolors='k')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title('PCA of Heart Disease Dataset')
plt.colorbar(label='Target')
plt.show()
```

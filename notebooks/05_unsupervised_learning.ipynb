# 05 - Unsupervised Learning using KMeans Clustering


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
# Load the cleaned dataset (after feature selection)
df = pd.read_csv("/content/Cleaned_dataset_AfterFeatureSelection.csv")
X = df.drop("target", axis=1)
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Apply KMeans clustering with 2 clusters
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_scaled)
```

```python
# Calculate the Silhouette Score to evaluate clustering performance
sil_score = silhouette_score(X_scaled, kmeans.labels_)
print("Silhouette Score:", round(sil_score, 3))
```

```python
# Plot the clustered data using the first two features
plt.figure(figsize=(10,6))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=kmeans.labels_, cmap='viridis', edgecolors='k')
plt.title("KMeans Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label='Cluster')
plt.show()
```
